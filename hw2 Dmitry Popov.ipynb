__Замечание.__ Можно считать, что все указанные объекты непустые (к примеру, в __задаче 1__ на диагонали матрицы есть ненулевые элементы) и корректные.

Невекторизованные реализации:

```python
def prod_non_zero_diag(m):
    res = 1
    for i in range(min(len(m), len(m[0]))):
        if abs(m[i][i]) > 1e-16:
            res *= m[i][i]

    return res


def are_multisets_equal(x, y):
    return sorted(x) == sorted(y)


def max_after_zero(x):
    res = float('-inf')
    for i in range(1, len(x)):
        if abs(x[i - 1]) < 1e-16 and x[i] > res:
            res = x[i]

    return res


def convert_image(x, y):
    h = len(x)
    w = len(x[0])
    c = len(x[0][0])

    res = [[None] * w for i in range(h)]
    for i in range(h):
        for j in range(w):
            temp = 0
            for k in range(c):
                temp += x[i][j][k] * y[k]
            res[i][j] = temp

    return res


def run_length_encoding(x):
    if len(x) == 0:
        return [], []

    run_len = [1]
    numbers = [x[0]]

    for i in range(1, len(x)):
        if x[i] != x[i-1]:
            run_len.append(1)
            numbers.append(x[i])
        else:
            run_len[-1] += 1

    return numbers, run_len


def pairwise_distance(x, y):
    n = len(x)
    m = len(y)
    d = len(x[0])

    xi = [0] * n
    yi = [0] * m

    for i in range(n):
        for k in range(d):
            xi[i] += x[i][k] ** 2

    for i in range(m):
        for k in range(d):
            yi[i] += y[i][k] ** 2

    Z = [[xi[i] + yi[j] for j in range(m)] for i in range(n)]

    for i in range(n):
        for j in range(m):
            for k in range(d):
                Z[i][j] -= 2 * x[i][k] * y[j][k]
            Z[i][j] = abs(Z[i][j]) ** 0.5

    return Z

funcs = {}
funcs[0] = {'prod_non_zero_diag': prod_non_zero_diag,
           'are_multisets_equal': are_multisets_equal,
           'max_after_zero': max_after_zero,
           'convert_image': convert_image,
           'run_length_encoding': run_length_encoding,
           'pairwise_distance': pairwise_distance
           }
```

Векторизованные реализации:

```python
import numpy as np


def prod_non_zero_diag(m):
    dia = np.diag(m)
    return np.product(dia[dia != 0])


def are_multisets_equal(x, y):
    return np.all(np.sort(x) == np.sort(y))


def max_after_zero(x):
    ids = np.where(x == 0)[0] + 1
    if x[-1] == 0:
        ids = ids[0:-1]
    return np.max(x[ids])


def convert_image(x, y):
    return np.inner(x, y)


def run_length_encoding(x):
    if len(x) == 0:
        return [], []

    changes = np.where(x[1:] != x[:-1])[0]
    ids = np.concatenate(([0], changes + 1))
    ids2 = np.concatenate((ids, [len(x)]))
    run_len = ids2[1:] - ids2[:-1]

    return x[ids], run_len


def pairwise_distance(x, y):
    x = x.astype(float)
    y = y.astype(float)
    xyt = np.matmul(x, y.T)
    xs = (x ** 2).sum(axis=1)
    ys = (y ** 2).sum(axis=1)
    return np.sqrt(np.abs(xyt * -2 + xs[:,None] + ys[None,:]))


funcs[1] = {'prod_non_zero_diag': prod_non_zero_diag,
           'are_multisets_equal': are_multisets_equal,
           'max_after_zero': max_after_zero,
           'convert_image': convert_image,
           'run_length_encoding': run_length_encoding,
           'pairwise_distance': pairwise_distance
           }
```

__Ответ:__ Для признака LSTAT я подобрал степенную зависимость. Улучшить качество преобразованием других признаков не удалось. $R^2$ достигнут 0.75.

Помимо преобразований отдельный признаков полезными бывают их попарные взаимодействия. Воспользуйтесь PolynomialFeatures, чтобы добавить попарные произведения и квадраты всех признаков. Обучите Ridge-регрессию (подберите гиперпараметр!) и посчитайте $R^2$ на тесте. Сильно ли изменилось качество?

Для одной комбинаций метода масштабирования и линейной регрессии постройте зависимость метрики $R^2$ на кросс-валидации (`GridSearchCV.cv_results_['mean_test_score']`) от значения параметра регуляризации для различных степеней полиномов в `PolynomialFeatures`.

```python
#  Your code here
# ...


poly_ridges = [Pipeline([('poly', PolynomialFeatures(degree=d)), ('scaler', MinMaxScaler((-1, 1))), ('model', Ridge())]) 
               for d in range(2, 6)]

poly_ridges_gs = [GridSearchCV(poly_ridges[d-2], {'model__alpha': params}, return_train_score=True).fit(nx_train, y_train)
                 for d in range(2, 6)]
```

```python
plt.subplots(2, 2, figsize=(22, 16))

for i, model in enumerate(poly_ridges_gs):
    plt.subplot(2, 2, i + 1)
    plt.title(f'Качество гребневой регрессии при степени полиномов {i+2}', fontsize=20)
    plt.semilogx(params, model.cv_results_['mean_train_score'], 'g', label='train')
    plt.semilogx(params, model.cv_results_['mean_test_score'], 'r', label='test')
    plt.xlabel(r'Значение гиперпараметра $\alpha$', fontsize=18)
    plt.ylabel(r'Коэффициент детерминации ($R^2$)', fontsize=18)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.legend(loc='lower left', prop={'size': 16})
    plt.ylim((0, 1))

plt.show()
```

При каких значениях происходит переобучение? А недообучение? Почему?

__Ответ:__ Как можно видеть, оптимальное значение степени полиномов равно 3. До этого значения модель недообучена, так как не учитывает полностью взаимодействия между признаками, после него она начинает сильно переобучаться - если регуляризатор не используется, то метрика $R^2$ вообще становится отрицательным. Это происходит потому, что модель пытается слишком тонко настроиться под обучающую выборку и теряет обобщающую способность.
